As apart of this task, we'll transmute our target variable from the variables of "no" and "yes" to 0 and 1 respectively. We'll do this and view a sample of the output to confirm that we see numeric values after applying these steps.

Excellent! We now see numeric values in this variable. We can move on to creating dummy variables for the rest of our categorical variables and replacing the previous variables with the new dummy variables.

```python
# create dummy variables
for col in Churn.columns[1:]:
    attName = col
    dType = Churn[col].dtype
    missing = pd.isnull(Churn[col]).any()
    uniqueCount = len(Churn[attName].value_counts(normalize=False))
    # discretize (create dummies)
    if dType == object:
        Churn = pd.concat([Churn, pd.get_dummies(Churn[col], prefix=col)], axis=1)
        del Churn[attName]
```

For this task, we'll need to begin by pulling in the appripriate functions from sci-kit learn. Next we'll go ahead with building and validating our model. For a decision tree model, we do not need to scale our features as mentioned on page 96 of the course text*. 

For our alterations of at least two default arguments, we'll set criterion to "entropy" and max depth to 5. The defaults for these parameters are "gini" and "none".

*<em> Raschka, S., & Mirjalili, V. (2019). Python machine learning. Birmingham, England: Packt Publishing.</em>

```python
# invoke needed fucntions
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

# build our model
tree_model = DecisionTreeClassifier(criterion='entropy',
                                    max_depth=5,
                                    random_state=1)

# train our model
tree_model = tree_model.fit(X_train, y_train)
```

We'll use the next two cells to prepare and display a visual of the decision tree.

```python
# invoke function from library
from IPython.display import Image

# create dot_data variable with appropriate parameters
dot_data = tree.export_graphviz(tree_model, out_file=None, 
                         filled=True, rounded=True,  
                         special_characters=True)
```

```python
# invoke library
import pydotplus 

# create graph variable and display image
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png()) 
```

The decision tree in the output above is an image and can be zoomed into if downloaded or opened in another tab. This image of the tree helps us to understand where and how each split has occured.

Next, we'll validate our model.

```python
# validate model by predicting against test data
target_predicted_dt = tree_model.predict(X_test)
```

In order to see how well the decision tree model preformed, we'll need to import a few more functions and create our confusion matrix along with other metrics for output.

```python
# invoke functions from libraries
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import confusion_matrix, classification_report

# print metrics
print("DT Accuracy Score", accuracy_score(y_test, target_predicted_dt))
print(classification_report(y_test, target_predicted_dt))
print(confusion_matrix(y_test, target_predicted_dt))

# extract true_positives, false_positives, true_negatives, false_negatives
tn, fp, fn, tp = confusion_matrix(y_test, target_predicted_dt).ravel()
print("True Negatives: ",tn)
print("False Positives: ",fp)
print("False Negatives: ",fn)
print("True Positives: ",tp)
```

From this output we can observe that our decision tree model had an overall accuracy of 93.8%. The remaining metrics can be helpful for model comparison, however, we'll hold discussion until the later task of model comparison.

<font color='#2B3856'><b>h. </b>Cross validate your decision tree using CV= 10.</font>

```python
# invoke function from library
from sklearn.model_selection import cross_val_score

# verify decisions tree with 10-fold cross validation
scores = cross_val_score(tree_model, X_train, y_train, cv=10)
print("Cross Validation Score for each K",scores)
scores.mean()
```

Compared to the overall accuracy of 93.8%, the mean of our cross validation scores was 93.46%. These number's being close helps us to know that neither overfitting or underfitting are a concern for this model as it preformed consistantly across each fold.

However, it's important in the field of data science that we make sure we are addressing the correct business implications. When it comes to churn, the business is likely interested in a model to predict customer churn to leverage as a tool to reduce customer churn.

If the goal of our predictive modeling is to correctly identify customers which are likely to churn so that the business can make an effort to intervene and hopefully retain these customers, overall accuracy is not as important as reducing false negatives. To make sense of this, let's briefly describe the business context of each quartile of our confusion matrix.

- True Negatives: Customers who were predicted not to churn and in fact did not churn.
- False Positives:  Customers who were predicted to churn, but did not actually churn.
- False Negatives:  Customers who were predicted not to churn, but did in fact churn.
- True Positives:  Customers who were predicted to churn and in fact did churn.

Let's assume that whatever effort the business plans to make to retain potentially churning customers has a fixed cost, that this cost is not greater than the value of each customer, and the intervening effort is mostly effective at retaining customers at risk for churn. For those customers that were correctly predicted to not churn, there is no action or expenditure for the business. For those customers that were correctly predicted to churn, under our assumptions the business is able to mitigate loss by intervening and reducing churn. For those costomers who are predicted to churn but would actually not, this cost expenditure and action taken by the business may not be a complete waste as it could possibility result in strengthening the relationship with that customer. 

Say for example, the cost expenditure was a discount and the action was a call from a manager to check in with the customer and inform them of the discount. This is unlikely to make a false positive turn into a true positive. However, in the instances of our false negatives where a customer was incorrected predicted to not churn but infact did churn, the business suffers an unmitigated lost as there was no opportunity identified to allow for any attempted intervention. Because of this, it is in the business' best interest to minimize the ammount of false positives.

Let's now look at the total false negatives of each model:

- Decision Tree: 62
- KNN: 198 
- Logistic Regression: 168

If the best benefit to the business is to optimize for the reduction of customer churn, the decision tree classifier is the most effective at meeting the business' goal. Because this classifier has the best overall accuracy, and appears to be most effective at achieviing the percieved business goal, this is the most effective classifier.
